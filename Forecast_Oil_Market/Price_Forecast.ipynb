# Избавляемся от выбросов в данных

```python
def clean_outliers(series):
    if not np.issubdtype(series.dtype, np.number):
        return pd.Series([True] * len(series), index=series.index)
    
    q1, q3 = np.percentile(series.dropna(), [25, 75])
    IQR = q3 - q1
    lower = q1 - 1.5 * IQR
    upper = q3 + 1.5 * IQR
    return (series >= lower) & (series <= upper)

# mask dataframe with True = OK, False = outlier
mask = df.apply(clean_outliers)

# удаляем строки, где выбросов >= 10
df_clean = df[mask.sum(axis = 1) > 11].copy()
df_clean
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

y = df_clean[df_clean.columns[10:13]]
X = df_clean.drop(df_clean.columns[10:13], axis = 1)
split_entry = int(len(X) * 0.7)
# Не используем случайное разделение на обучаемую и тестовую выборку train_test_split, тк в обучаемую выорку могут попасть данные из будущего
X_train, X_test, y_train, y_test = X.iloc[:split_entry], X.iloc[split_entry:], y.iloc[:split_entry], y.iloc[split_entry:]
print(y_test)

X_train['month'] = X_train['Месяц'].dt.month
X_train['year'] = X_train['Месяц'].dt.year
X_train['days_in_month'] = X_train['Месяц'].dt.days_in_month

X_test['month'] = X_test['Месяц'].dt.month 
X_test['year'] = X_test['Месяц'].dt.year
X_test['days_in_month'] = X_test['Месяц'].dt.days_in_month

X_to_forecast['month'] = X_to_forecast['Месяц'].dt.month 
X_to_forecast['year'] = X_to_forecast['Месяц'].dt.year
X_to_forecast['days_in_month'] = X_to_forecast['Месяц'].dt.days_in_month

# Удалить уже лишнюю колонку даты
X_train = X_train.drop('Месяц', axis=1)
X_test = X_test.drop('Месяц', axis=1)
X_to_forecast = X_to_forecast.drop('Месяц', axis = 1)

new_columns = X_train.columns
X_train.info()

# Масштабиорвагние признаков
standartScaler = StandardScaler()
standartScaler.fit(X_train)
X_train = standartScaler.transform(X_train)
X_test = standartScaler.transform(X_test)
X_to_forecast = standartScaler.transform(X_to_forecast)
X_train.shape

```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

lasso = Lasso()
grid = {'alpha': [0.01, 0.1, 0.5, 1, 5, 5.5, 5.3, 10, 100]}

gridSearch = GridSearchCV(lasso, param_grid=grid, cv=5, scoring='neg_mean_squared_error')
gridSearch.fit(X_train, y_train)

print("Лучший alpha:", gridSearch.best_params_)

coeffs = gridSearch.best_estimator_.coef_
print("Коэффициенты:", coeffs)

# Предсказание
y_predict = gridSearch.predict(X_test)
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_predict))}, R2: {r2_score(y_test, y_predict)}')


for i, coef_vector in enumerate(coeffs):
    print(f"\nКоэффициенты для y[{i}]:")
    for name, coef in zip(new_columns, coef_vector):
        print(f"{name}: {coef:.4f}")
```
